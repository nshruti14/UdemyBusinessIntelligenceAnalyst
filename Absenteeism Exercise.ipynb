The goal is to predict absenteeism from work

```python
import pandas as pd
```

```python
df=pd.read_csv('Absenteeism_data.csv')
df.head()
```

Absenteeism Time in Hours is our dependent variable

```python
#Make a copy of the data frame
df_copy=df.copy
```

```python
df.info()
```

```python
#No missing values
#Drop variables that we don't need
df.drop('ID', axis=1, inplace=True)
#df no longer has column ID
df
```

```python
#Take a deeper look at the Reason for Absence column
sorted(df['Reason for Absence'].unique())
```

Notice that the number 20 is missing. The "Reason for Absence" column contains integers that are substituted for reasons (It is categorical).

```python
#Let's convert the categorical variable into dummy variables
reason_columns=pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
#Let's add a row called check to see if there is only "1" in every row
reason_columns['check']=reason_columns.sum(axis=1)
reason_columns
```

```python
#What if all columns don't have the value 1? Let's check
reason_columns['check'].unique()
```

```python
#Now we can be sure that all values in the check column are 1. We can now drop the check column
reason_columns.drop('check', axis=1, inplace=True)
reason_columns.head(1)
```

```python
#Drop the first column from reason_columns since it acts as the baseline (worker is away for an unknown reason)
reason_columns=pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns.head(1)
```

```python
#There are 28 different reasons for absence, can we group them to lower the number?
df.drop('Reason for Absence', axis=1, inplace=True)
reason_columns.columns.values
```

```python
#These are grouped based on the list of reasons provided
reason_type1=reason_columns.loc[:, 1:14].max(axis=1)
reason_type2=reason_columns.loc[:, 15:17].max(axis=1)
reason_type3=reason_columns.loc[:, 18:21].max(axis=1)
reason_type4=reason_columns.loc[:, 22:].max(axis=1)
```

```python
#Let's add the 4 reason types to the DataFrame
df=pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis=1)
df
```

```python
#Let's change the names of the columns
df.columns.values
```

```python
column_names=['Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns=column_names
```

```python
df.head(1)
```

Reorder the columns so the reasons for absence appear at the beginning

```python
column_names_reordered=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
```

```python
df=df[column_names_reordered]
df.head(1)
```

```python
#We now have the reasons for absence at the beginning like it was before we manipulated the column
#Let's create a checkpoint here in case we need to come back to this point without restarting from the beginning
df_reason_mod=df.copy()
```

Analyze the date column

```python
#Check current type of data in the Date column
type(df_reason_mod['Date'][0])
```

```python
#We want the data to be timestamps
df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'], format='%d/%m/%Y')
type(df_reason_mod['Date'][0])
```

```python
#Let's extract the month value from the date column to see if employees tend to be more absent in certain months 
list_months=[]
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month value']=list_months
```

```python
df_reason_mod.head()
```

```python
#Let's extract the day of the week using a function
def day_of_week(date_value):
    return date_value.weekday()
df_reason_mod['Day of Week']=df_reason_mod['Date'].apply(day_of_week)
```

```python
df_reason_mod.head()
```

```python
#Make a copy of df before dropping the Date column
df_reason_mod_with_date=df_reason_mod.copy()
```

```python
#We can now drop the date column and reorder the columns so Month and Day of the Week appear where Date used to be
df_reason_mod_no_date=df_reason_mod_with_date.drop(['Date'], axis=1)
```

```python
df_reason_mod_no_date.columns.values
```

```python
columns_reord=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month value', 'Day of Week', 'Transportation Expense', 'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
df_reason_mod_no_date=df_reason_mod_no_date[columns_reord]
df_reason_mod_no_date.head()
```

Analyze the Education column

```python
df_reason_mod_no_date['Education'].unique()
```

```python
df_reason_mod_no_date['Education'].value_counts()
```

We are told that the values represent the following: 1 represents high school, 2 represents graduate, 3 represents postgraduate, and 4 represents a master or a doctor.

Since a majority of the people have a high school education, separating for the other levels of education becomes less relevant. We can merge the education values 2, 3, and 4.

```python
#Map the values in the Education column using a dictionary
df_reason_mod_no_date['Education']=df_reason_mod_no_date['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
#Check to see if Education values were mapped correctly
df_reason_mod_no_date['Education'].unique()
```

```python
df_reason_mod_no_date['Education'].value_counts()
```
